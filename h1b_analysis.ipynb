## Dropping irrelevant columns based on either their value types or redundancy

```python
columns_to_drop = ['CASE_NUMBER', 'JOB_TITLE', 'TRADE_NAME_DBA', 'EMPLOYER_NAME',
       'TRADE_NAME_DBA', 'EMPLOYER_ADDRESS1', 'EMPLOYER_ADDRESS2',
       'EMPLOYER_CITY','EMPLOYER_POSTAL_CODE',
       'EMPLOYER_COUNTRY', 'EMPLOYER_PROVINCE', 'EMPLOYER_PHONE',
       'EMPLOYER_PHONE_EXT', 'AGENT_ATTORNEY_LAST_NAME', 'AGENT_ATTORNEY_FIRST_NAME',
       'AGENT_ATTORNEY_MIDDLE_NAME', 'AGENT_ATTORNEY_ADDRESS1',
       'AGENT_ATTORNEY_ADDRESS2', 'AGENT_ATTORNEY_CITY',
       'AGENT_ATTORNEY_STATE', 'AGENT_ATTORNEY_POSTAL_CODE',
       'AGENT_ATTORNEY_COUNTRY', 'AGENT_ATTORNEY_PROVINCE',
       'AGENT_ATTORNEY_PHONE', 'AGENT_ATTORNEY_PHONE_EXT',
       'AGENT_ATTORNEY_EMAIL_ADDRESS', 'LAWFIRM_NAME_BUSINESS_NAME',
       'STATE_OF_HIGHEST_COURT', 'NAME_OF_HIGHEST_STATE_COURT',
       'WORKSITE_WORKERS',
       'SECONDARY_ENTITY_BUSINESS_NAME', 'WORKSITE_ADDRESS1',
       'WORKSITE_ADDRESS2', 'WORKSITE_CITY', 'WORKSITE_COUNTY',
       'WORKSITE_POSTAL_CODE', 'WAGE_RATE_OF_PAY_FROM',
       'WAGE_RATE_OF_PAY_TO', 'WAGE_UNIT_OF_PAY', 'PREPARER_LAST_NAME',
       'PREPARER_FIRST_NAME', 'PREPARER_MIDDLE_INITIAL',
       'PREPARER_BUSINESS_NAME', 'PREPARER_EMAIL', 'EMPLOYER_POC_LAST_NAME',
       'EMPLOYER_POC_FIRST_NAME', 'EMPLOYER_POC_MIDDLE_NAME',
       'EMPLOYER_POC_JOB_TITLE', 'EMPLOYER_POC_ADDRESS1',
       'EMPLOYER_POC_ADDRESS2', 'EMPLOYER_POC_CITY', 'EMPLOYER_POC_STATE',
       'EMPLOYER_POC_POSTAL_CODE', 'EMPLOYER_POC_COUNTRY',
       'EMPLOYER_POC_PROVINCE', 'EMPLOYER_POC_PHONE', 'EMPLOYER_POC_PHONE_EXT',
       'EMPLOYER_POC_EMAIL']
```

```python
h1b_filtered = h1b.drop(columns=columns_to_drop,axis=1)
```

## columns with a lot of nulls  can also be dropped since they are redundant

```python

redundant_columns = ['ORIGINAL_CERT_DATE','PW_TRACKING_NUMBER','PW_OTHER_SOURCE','PW_OTHER_YEAR','PW_SURVEY_PUBLISHER','PW_SURVEY_NAME','WILLFUL_VIOLATOR', 'SUPPORT_H1B','STATUTORY_BASIS','APPENDIX_A_ATTACHED']
```

```python
h1b_filtered = h1b_filtered.drop(columns=redundant_columns,axis=1)
```

# SOC_TITLE gives a broader idea of the job title.However,one title i.e. software engineer appears a very high amount of times. Due to this, we can filter our data based on the value counts of each title. A good idea could be to consider only the titles that have a value count between 100 - 10000

```python
df = h1b_filtered[(h1b_filtered['SOC_TITLE'].map(h1b_filtered['SOC_TITLE'].value_counts()>100)) & (h1b_filtered['SOC_TITLE'].map(h1b_filtered['SOC_TITLE'].value_counts()<10000))]
```
